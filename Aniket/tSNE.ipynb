```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.manifold import TSNE
```

```python
df = pd.read_csv("Encoded_Data.csv")
```

```python
df.head()
```

```python
df = df.fillna(method='ffill')
```

```python
X = df.drop(['Economic Need Index', 'Address (Full)', 'School Income Estimate', 'Location Code'], axis = 1)
X = df.drop(df.columns[:38], axis = 1)
```

```python
tsne = TSNE(n_components=3, verbose = 2)
```

```python
tsne_out = tsne.fit_transform(X)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(tsne_out[:,0], tsne_out[:,1], c = df['Economic Need Index'])
plt.xlabel("TSNE1")
plt.ylabel('TSNE2')
plt.colorbar()
```
